```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from dataset import *
from pprint import pprint
```

```python
config = {
    'dataset': 'Lsun_church',
    'data_path': '/root/notebooks/data/church_outdoor_train_lmdb/',
    'data_size': 16,
    'use_image_generator': False,
    'batch_size': 16,
    'num_sample': 16
}
```

```python
ds, info = get_dataset_and_info(config)
bds = next(iter(ds))
samples = np.uint8(bds[0].numpy()*127.5 + 128).clip(0, 255)
```

```python
fig = plt.figure(figsize=(8, 8))
gs = gridspec.GridSpec(8, 8)
gs.update(wspace=0.05, hspace=0.05)

for i, sample in enumerate(samples):
    ax = plt.subplot(gs[i%4*2:(i%4*2+2), i//4*2:(i//4*2+2)])
    plt.axis('off')
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_aspect('equal')
    plt.imshow(sample)

plt.show()
```
